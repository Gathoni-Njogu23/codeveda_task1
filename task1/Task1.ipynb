## checking price validity
for this case low <= open,low<close ,open <=close ,close<=high


```python
invalid_prices = df[
    (df['open'] < df['low']) |
    (df['open'] > df['high']) |
    (df['close'] < df['low']) |
    (df['close'] > df['high'])
]
```

```python
invalid_prices.shape
```

```python
df = df.drop(invalid_prices.index)
```

```python
df.reset_index(drop=True, inplace=True)
```

for volume since we have very high spikes hence having very high extreme values i decided to use the IQR then cap them so that they can remain within their range


```python
df['volume'].describe()
```

```python
Q1 = df['volume'].quantile(0.25)
Q3 = df['volume'].quantile(0.75)
IQR = Q3 - Q1
```

```python
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
```

for outliers above high should be above 50 million and below low is about 500,000

```python
df_copy = df.copy()
df_copy['volume_winsorized'] = np.where(df_copy['volume'] > 50000000, 50000000,
                                        np.where(df_copy['volume'] < 500000, 500000, df_copy['volume']))
```

```python
df = df.copy()  # now df is independent
df['volume_winsorized'] = df_copy['volume_winsorized']
```
